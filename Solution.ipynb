```python
# Loading libraries
import pandas as pd
import geonamescache
import re
```

```python
# Reading file
text = open("data/headlines.txt").read()
```

```python
# Getting headlines
headlines = text.split("\n")
headlines[0]
```

```python
gc = geonamescache.GeonamesCache()
countries = gc.get_countries()
# print countries dictionary
names = countries.get('AD')['name']
print(names)
```

```python
# Getting countries names
# Create empty list
lst_countries = []
# Filling list with countries names
for key in countries.keys():
    lst_countries.append(countries.get(key)['name'])
print(lst_countries)
```

```python
# Getting cities names
cities = gc.get_cities()
cities_names = cities.get('3041563')['name']
print(cities_names)
```

```python
# Getting all cities names
# Create empty list
lst_cities = []
for key in cities.keys():
    lst_cities.append(cities.get(key)['name'])
print(lst_cities)
```

```python
regex = 'Recife'
random_text = 'Clown Patty'
match = re.search(regex, random_text)
print(match)
#assert match is None
```

```python
matchable_text = 'Mystery Virus Spreads in Recife, Brazil'
match = re.search('Bra', matchable_text)
if match:
    print('lo encontro')
else:
    print('no lo encontro')
#assert match is not None
start, end = match.start(), match.end()
matched_string = matchable_text[start: end]
print(matched_string)
#assert matched_string == 'Boston'
```

```python
pais = []
for text in headlines:
    flag = 'NaN'
    for country in lst_countries:
        match = re.search(country,text)
        if match:
            flag = country
            pais.append(country)
    if flag == 'NaN':
        pais.append(flag)
print(pais)
```

```python
ciudad = []
for text in headlines:
    flag = 'NaN'
    for city in lst_cities:
        res = re.findall(r'\w+',text)
        #match = re.search(regex,text)
        if city in res and flag == 'NaN':
            flag = city
            ciudad.append(city)
    if flag == 'NaN':
        ciudad.append(flag)
print(ciudad)
```

```python
titular = pd.DataFrame(headlines)
print(titular)
```

```python
country = pd.DataFrame(pais)
print(country)
```

```python
city = pd.DataFrame(ciudad)
print(city)
```

```python
data = pd.concat([titular,country,city],axis=1)
data.columns = ['headlines','countries','cities']
print(data)
```
